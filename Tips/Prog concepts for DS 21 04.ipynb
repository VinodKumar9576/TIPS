#Recursion/Memorization

* Recursion is a top-down approach.

**Factorial using recursion**

```
def factorial(n):
  if n==0:
    return 1
  return n*factorial(n-1)
```

**Fibonacci number using recursion**
<br>Series : 1,2,3,5,8,13,.....

```
#finding nth fibonacci number
def fib(n):
  if n<=1:
    return 1
  return fib(n-1) + fib(n-2)
```

* Above recursion problem calculates n-th numbers repetitively for larger fibonacci numbers.
* Now we are going to find a way to avoid repetitive calculations.
* To store the older calculations we use dictionary.

**Repititive calculation**
<br>
image.png

```
memo = {}

def fib_memo(n):
  if n in memo:
    return memo[n]
  if n<=1:
    memo[n]=1
    return 1
  memo[n] = fib_memo(n-1)+fib_memo(n-2)
  return memo[n]
```

#Dynamic programming
* Bottom-up approach
* A way to write recursion iteratively.
* We start calculating fib(0) and fib(1) and then use previous results to generate new results.

```
def fib_dp(n):
  dp_sols = {0:1,1:1}
  for i in range(2,n+1):
    dp_sols[i] = dp_sols[i-1] + dp_sols[i-2]
  return dp_sols[n]
```

* Both are linear but DP method is faster than memorization method.
* Coz DP method made call exactly 1 time to each subproblem in this case. 

#Binary Search
* Sorted array of elements
* Obj: to search an element
* Advanced case of recursion based algorithm.
* Recursively looking at the middle element and see if the search element is left or right side of the middle element.
* Search space reduced by factor of 2 each step.
* R.T.C : O(logn)

**Comparision b/w linear search & Binary Search**

image.png

```
#Returns indexes of target in nums array if element is present, else -1

def binary_search(nums,left,right,target):

  #Base code
  if right>=left:
    mid = (right+left)//2
    #if target is at mid return
    if nums[mid]==target:
      return mid
    
    elif nums[mid]>target:
      #searching in right search space
      return binary_search(nums,left,mid-1,target)
    
    else:
      #searching in left search space
      return binary_search(nums,mid+1,right,target)
  else:
    #target is not in given numbers
    return -1

nums = [1,2,3,4,5,6,7,8,9]
print(binary_search(nums,0,len(nums)-1,7))
```
