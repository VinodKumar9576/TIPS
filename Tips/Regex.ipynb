```
# Finding number of punctuations in string without regex module.
target = [';',':',',','.','-','\'']
stri = "It was the best of times, it was the worst of times, it was the age of wisdom, it was the age of foolishness, it was the epoch of belief, it was the epoch of incredulity, it was the season of Light, it was the season of Darkness, it was the spring of hope, it was the winter of despair, we had everything before us, we had nothing before us, we were all going direct to Heaven, we were all going direct the other way – in short, the period was so far like the present period, that some of its noisiest authorities insisted on its being received, for good or for evil, in the superlative degree of comparison only.**"

count=0
for punc in target:
  if punc in stri:
    count+=stri.count(punc)
print(count)
```

```
# With regex

import re

pattern = r'[,;.,–]'

stri = "It was the best of times, it was the Ddb85ee worst of times, it was the age of wisdom, it was the age of foolishness, it was the epoch of belief, it was the epoch of incredulity, it was the season of Light, it was the season of Darkness, it was the spring of hope, it was the winter of despair, we had everything before us, we had nothing before us, we were all going direct to Heaven, we were all going direct the other way – in short, the period was so far like the present period, that some of its noisiest authorities insisted on its being received, for good or for evil, in the superlative degree of comparison only.**"

print(len(re.findall(pattern,stri)))
print(re.findall(pattern,stri))
```

## 1. the [] operator

* [abc] -  will find all occurances of a or b or c
* [a-z] - will find all occurances of lowercase a-z
* [a-zA-Z0-9] - will find all occurances of a-z, A-Z, 0-9.

```
pattern = r'[a-zA-Z]'

print(len(re.findall(pattern,stri)))
```

## 2. The dot Operator

* The dot operator is used to match single instance of any character excpt new line character.
* We can use dot operator in conjunction with another.

```
 # Find the substrings in string that start with small 'd' or capital 'D'
 # and end with 'e' with a lenght of 6.

pattern = r'[dD]....e'
# as length is 6 chars, we gave first letter as d or D and 4 dots then last letter e

print(re.findall(pattern,stri))
```

## 3. Some Meta Sequences

* \w - Matches any letter, digit or underscore. Equivalent to [a-zA-Z0-9_]
* \W - Matches anything other than letter, digit or under score
* \d - Mathces any decimal digit. Equivalent to [0-9]
* \D - Matches anything other than a decimal digit.

## 4. The Plus and Star operator

* Plus + char is used to signify one or more instance of leftmost char.
* Star * char is used to signify 0 or more instance of leftmost char.

```
# To find substring of string that starts with d or D and end with e with atleast one char
# b/w d and e. d\w*e

pattern = r'[dD]\w*e'
# To have words that starts with d/D end with e and have atleast one char.
# Returns the words that matched the given pattern.
# pattern = r'[dD]\w+e'

print(re.findall(pattern,stri))
```

* **More generic approach**
  * Instead of * or + i.e 0 or more times or 1 or more times we use specific n times.
  * \w{n} - Repeat \w exactly n times
  * \w{n,} - Repeat \w atleast n times or more
  * \w{n1, n2} - Repeat \w n1 times not more than n2 times

```
pattern = '[dD]\w{3}e'
#prints words that have two characters in b/w d/D and e
# returns the specified matched pattern from any word and not whole word.
# below is returned from 'Ddb85ee'
print(re.findall(pattern,stri))
```

## 5. ^ Caret Operator and $ Dollar operator.

* '^' matches the start of the string '$' matches the end of string

```
pattern = r'time$'
print(re.findall(pattern,stri))
```

* In all above examples always a substring is returned not the whole word.
* Now we want to find all words that match the given pattern.

## 6. Word Boundary

```
# Lets say we want to find words that starts with d.

pattern = r'd\w*'

print('O/p of pattern : \'d\w*\'')
print(re.findall(pattern,stri))
# from above pattern you get substring of words as seen in output.
# But we want only the exact words.

pattern = r'\bd\w*'
print('='*100)
print('O/p of pattern : \'\\bd\w*\'')
print(re.findall(pattern,stri))
```

# Regex Functions

## 1. findall

* Input: Pattern,string;  Output: List of strings

```
pattern = r'[iI]t'
str = "It was the best of times, it was the worst of times."

print(re.findall(pattern,str))
```

## 2. Search

* Input: Pattern,string;  Output: Location object for the first match

```
location = re.search(pattern,str)
print(location)

# we can get the data of this location
print(location.group())
```

## 3. Substitute

* Input: Search Pattern,Replacement pattern, Test string; Output: Substituted string

```
# Replace the word 'time' with 'life'

string = re.sub(r'times',r'life',str)
# in above pattern the exact word is replaced with replacement string.
print(string)
```

# Some Case Studies:

**Finding Pan Numbers**

```
pattern = r'\b[A-Z]{5}[0-9]{4}[A-Z]'

string ="ABcDE1234L"

match = re.search(pattern,string)

if match:
  print(True)
else:
  print(False)
```

**Finding Domain Names**

```
pattern = r'[w]{3}\.[a-z]+\.\w{2,3}'
string = "www.google.com"

match = re.search(pattern,str)
print(match.group())
print(re.findall(pattern,string))

if match:
  print(True)
else:
  print(False)
```
