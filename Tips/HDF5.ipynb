**Important notes:**

* HDF5 file stands for Hierarchical Data Format 5. 
* It stores data in a hierarchical structure within a single file.
* So if we want to quickly access a particular part of the file rather than the whole file, we can easily do that using HDF5.
* This functionality is not seen in normal text files hence HDF5 is becoming seemingly popular in fact of being a new concept.

Task 1: Create a file and save a random array of numpy in it.

```
import numpy as np
import h5py

#initializing a random numpy array
arr = np.random.randn(1000)

#creating a file
with h5py.File('/content/drive/My Drive/Colab Notebooks/Tips/test.hdf5', 'w') as f:
  dset = f.create_dataset('default', data = arr)
# 'default' is the dataset name that's going to be stored in file test.hdf5   
```

1. We open the file as “write only” attribute.<br>This means that if there
isn’t any file with the name test.hdf5 then it will create one otherwise it will delete (overwrite) the content of the existing file.<br>
2. While opening the file, we used "with" instead of "open" as it has an upper-hand when compared to open() method. We don’t need to close a file if we open it using "with".<br>
3. Finally, we use the .create_dataset() to set the varibale dset to the array which was created earlier.


Task 2:  Read the file which we wrote above:

```
#open the file as f
with h5py.File('/content/drive/My Drive/Colab Notebooks/Tips/test.hdf5', 'r') as f:
  data = f['default']

  #get the minimum value
  print(min(data))

  # get the maximum value
  print(max(data))

  #print few values
  print(data[:15])

print(type(data))
```

1. We set the variable data to the data we stored in the previous file.
2. But, the variable data is not an array. It’s a dataset. Rather than storing data in the RAM, it saves it in hard-drive of the computer, thus maintaining a hierarchy of structures just like directories:
