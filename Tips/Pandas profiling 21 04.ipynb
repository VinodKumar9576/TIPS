> **Quick EDA : Pandas Profiling.**

**EDA Cycle**
* Understanding Data Quality
* Description
* Shape
* Patterns
* Relationships
* Visualizing the data

For a given dataset, pandas profiling computes the following:
* Essentials :
  * type
  * unique values
  * missing values
* Quantile statistics :
  * minimum value
  * Q1
  * median
  * Q3
  * maximum range
  * Interquartile range
* Descriptive statistics:
  * Mean
  * Mode
  * Standard Deviation
  * Sum
  * Median Absolute Deviation
  * Coefficient of variation
  * Kurtosis
  * Skewness
* Most Frequent values
* Histogram
* Correlations:
  * Correlated variables
  * Spearman
  * Pearson matrices
* Sample of dataset
