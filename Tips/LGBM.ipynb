###**Introduction**

* Light GBM is a gradient boosting framework that uses tree based learning algorithm.
* It grow tree vertically focussing on single node at a time, where as other algorithms grow horizontally by level-wise.
* Algo choose the leaf with **maximum delta loss** to grow.
* WHen growing the same leaf, Leaf-wise algorithm can reduce more loss than a leve-wise algorithm.

###**Specialities that make Light GBM more popular**

* LGBM can handle the large size of data and takes **lower memory to run**.
* Another perk is that LGBM focuses on **accuracy of results.**
* LGBM also **supports GPU learning** and thus data scientists are widely using LGBM for data science applications development.

###**Where can we use LGBM**

* Not for every problem we can use LGBM as its not advisable to use LGBM on small datasets.
* LGBM is sensitive to overfitting and can easily overfit small data.
* There's no thumb rule on data set size suitable for LGBM but it is advisable to use only for data with 10,000+ rows.

###**Parameters**

####**Control Parameters**

* **max_depth :**
  * Says maximum depth of tree.
  * used to handle model overfitting.
  * If you find model is overfit, lower this parameter.

* **min_data_in_leaf :**
  * Min no. of records a leaf may have.
  * Default value is 20, optimum value.
  * Also used to deal overfitting.

* **feature_fraction :**
  * Used when your boosting is random forest.
  * 0.8 feature fraction means that LGBM will select 80% of feautures randomly in each iteration for building trees.

* **bagging_fraction :**
  * Fraction of data to be used for each iteration and is generally used to speed up the trainning and avoid overfitting.

* **early_stopping_round :**
  * This can help speed up the analysis.
  * Model will stop training if one metric of one validation data doesn't improve in last early_stopping_round rounds.
  * This will excessive iterations.
  * Model will stop learning if over last early_stopping_round rounds, model learning didn't improve.

* **lambda :**
  * This specifies regularization.
  * Value ranges from 0 to 1.

* **min_gain_to_split :**
  * This parameter will describe the minimum gain to make a split.
  * Used to control no. of useful splits in a tree.

* **max_cat_group**
  * When number of category is large, finding the split point on it is easily over-fitting.
  * So LGBM merges them into 'max_cat_group' groups, and finds the split points on the group boundaries.
  * default: 64

####**Core Parameters**

* **Task :**
  * Specifies the task you want to perform on data.
  * It may be either train or predict.

* **application :**
  * This is the most important parameter and specifies the application of your model whether it is a regression problem or classification problem.
  * LGBM will by default consider model as a regression model.
  * **regression**: for regression
  * **binary** : for binary classification
  * **multiclass** : for multiclass classification problem

* **bossting :**
  * Defines the type of algorithm you want to run, default = gbdt
  * **gbdt** : traditional **Gradient Boosting Decision Tree**
  * **rf :** random forest
  * **dart :** Dropouts meet Multiple Additvive Regression Trees
  * **gross :** Gradient-based One-Side Sampling

* **num_boost_round :**
  * Number of boosting iterations, typically 100+

* **learning_rate :**
  * This determines the impact of each tree on the final outcome.
  * GBM works by starting with an initial estimate which is **updated using the output of each tree**.
  * The learning parameter controls the magnitude of this change in the estimates.
  * General values: 0.1, 0.001, 0.003..

* **num_leaves :**
  * Number of leaves in full tree.
  * Default : 31

* **device :**
  * default : cpu, can also pass gpu

####**Metric Parameters**

* **metric :**
  * One of the important parameter as it specifies loss for the model.
  * **mae :** mean absolute error
  * **mse :** mean squared error
  * **binary_logloss :** loss for binary classification
  * **multi_logloss :** loss for multi classification

####**IO Parameters**

* **max_bin :**
  *It denotes the **maximum number of bin that feature value** will bucket in.

* **categorical_features :**
  * It denotes the index of categorical features.
  * If categorical_features = 0, 1, 2 then column 0, column 1, column 2 are categorical variables.

* **ignore_column :**
  * Same as categorical_features just instead of considering specific columns as categorical, it will completely ignore them.

* **save_binary :**
  * If you are really dealing with the memory size of your data file then specify this parameter as 'True'.
  * Setting this to true will save the dataset to binary file.
  * This file will speed your data reading time for the next time.
