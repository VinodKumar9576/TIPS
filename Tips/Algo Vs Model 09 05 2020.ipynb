**<h1>Difference between algorithm and model**

**<h3>What is an 'Algorithm' in Machine Learning**

* An algorithm in machine learning is a ***procedure*** that is run on data to create machine learning model.
* Machine learning algorithms perform ***pattern recognition*** and can ***learn*** from data or ***fit*** on a dataset.
* Many machine learning algorithms are in place such as :
  * Classification algos. Eg: KNN
  * Regression. Eg: Linear regression
  * Clustering algo. Eg: K-means
* Properties of ML algos:
  * Can be described using math and pseudocode
  * Efficieny can be analyzed and described
  * Can be implemented with any of modern programming languages.
* Multiple algos can also be seen implemented together and provided in a library with standard application programming interface (API). Eg: scikit-learn

**<h3>What is a 'Model' in Machine Learning**

* Model is the output of a ML algorithm and represents what was learned by algo.
* Model is the ***entity*** that is saved after running a machine learning algorithm on training data and represents rules,numbers and any other algo-specific data structures required to make predictions.

> **Examples**
* Linear regression algorithm : results a model that has vector of coefficients for linear structure.
* Decision Tree : if-then statements with specific values.
* Neural Network / backpropagation / gradient descent algorithms : result in graph structure with vectors or matrices of weights.

* Think ML model as a ***Program***.
* Model contains both data and procedure for using the data to make prediction.
* Take linear regression : `y = mx + c`
  * Linear regression outputs model coefficients which we use those coeff to multiply and sum with row of new data to make predictions.<br>

**`Machine Learning Model == Model Data + Prediction Alogorithm`**

**<h3>Algorithm Vs Model Framework**

* We already got familiarized with ***Algorithm vs Model***
* Algorithms have all of their work in the ***algorithm*** and the ***prediction algorithm*** does very less.
* Algorithm is like optimization procedure that minimizes error of model(data + prediction) on training dataset.
  <br>Eg: *Linear Regression* performs optimization process to find a set of weights that minimize the sum squared error on the training dataset.

(Frame work)<br>
**Linear Regression**
  * **Algorithm**: Find set of coeff that minimize error on training dataset.
  * **Model**:
    * **Model Data**: Vector of coefficients.
    * **Prediction Algorithm**: Multiply and sum coefficients with input row.

* Some algorithms are trivial and do nothing as all of the work resides in model or prediction algorithm.

  **Eg**: KNN has no *`algorithm`* other than saving the entire training dataset. In this case *`model data`* is the entire training dataset and all of the work is in the prediction algorithm.

**K-Nearest Neighbors**
  * **Algorithm**: Save training data.
  * **Model**:
    * **Model Data**: Entire training dataset.
    * **Prediction Algorithm**: Find k most similar rows and average their target value(in case of regression), majority vote (in case of classification).

**<h3>Machine learning is Automatic Programming**

* We just want a machine learning *model* and *algorithm* is just the path we follow to get the model.
* ML techniquest are used for problems that can't be solved efficiently or effectively in other ways.

> **Eg:** To classify emails as spam or not.
* Writing code manually and reviewing tonnes of mails to achieve this is simply slow, fragile and not very effective.
* Instead we can use ML techniques to solve this like ***Naive Bayes***.
* We don't want *Naive Bayes* algorithm, we want the model that *Naive Bayes* results so that we can use it to classify emails (vectors of probabilites and prediction algorithm).

* So in this sense, the ML model is a program automatically written or created or learned by ML algorithm to solve our problem.
