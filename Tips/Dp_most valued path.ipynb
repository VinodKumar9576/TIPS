# Fibonacci Series by Recursion

```
def fib(n):
  if n<=1:
    return 1
  return fib(n-1) + fib(n-2)

# With above approach we have to multiple times computing fib(5),fib(4) .. etc.
# This is Top down approach, and if n is large, computation time increases.
# We go now by bottom Up approach with simple for loop
```

# Fibonacci series by Bottom Up approach

```
def fib_dp(n):
  dp_sols={0:1,1:1}
  for i in range(2,n+1):
    dp_sols[i]=dp_sols[i-1]+dp_sols[i-2]
  return dp_sols[n]
```

```
def maxPathSum(grid):
  m = len(grid) # no. of columns
  n = len(grid[0]) # No. of rows
  sol = list(grid)

# Filling first row
  for i in range(1,n):
    sol[0][i]+=sol[0][i-1]

# Filling first column
  for i in range(1,m):
    sol[i-1][0]+=sol[i-1][0]

# Filling rest cells based on max surrounding values
  for i in range(1,m):
    for j in range(1,n):
      sol[i][j]+= max(sol[i-1][j], sol[i][j-1])
# return the last element. Last cell in matrix holds max value you can get if you come in the effective path
  return sol[-1][-1]  
```
